```python
# import sys
# l = []
# print(sys.getsizeof(l))
# l.append(1)
# print(sys.getsizeof(l))
# l.append('a')
# for i in range(50):
#     print(i,sys.getsizeof(l))
#     l.append(i)

import ctypes #python library is used to create python list using c datatype array

class pList:
    def __init__(self):
        self.size = 1 # size of the initialize array
        self.n = 0  # no of element present in array and index no
        #create a C type array with size = self.size
        self.A = self.__make_array(self.size)

    def __make_array(self, capacity): #capacity means size of array
        #creates a C type array(static, referential) with size capcity
        return (capacity*ctypes.py_object)()

    def append(self,item):
        if self.n == self.size:
            #resize
            self.__resize(self.size*2)

        self.A[self.n] = item
        self.n = self.n + 1
    
    def __resize(self, new_capacity):
        #create a new array with new capacity
        B = self.__make_array(new_capacity)
        self.size = new_capacity
        #copy the content of A to B 
        for i in range(self.n):
            B[i] = self.A[i]
        #reassign A
        self.A = B

    def __len__(self): #magic method
        return self.n
    
    def pop(self):
        if self.n ==0:
            return 'Empty list'
        print(self.A[self.n-1])
        self.n = self.n - 1

    def clear(self):
        self.n = 0
        self.size = 1

    def find(self,item):
        for i in range(self.n):
            if self.A[i] == item:
                return i
        return 'ValueError: item not in the list'

    def insert(self,pos,item):
        if self.n == self.size:
            self.__resize(self.size*2)
        for i in range(self.n,pos,-1):
            self.A[i] = self.A[i-1]
        self.A[pos] = item
        self.n = self.n +1



    def __getitem__(self,index): # magic method
        if 0 <= index <= self.n:
            return self.A[index]
        else:
            return 'IndexError: index out of range'
        

    def __delitem__(self,pos):
        if 0 <= pos <= self.n:
            for i in range(pos, self.n-1):
                self.A[i] = self.A[i+1]
            self.n = self.n - 1
        else:
            return 'IndexError: index out of range'


    def remove(self,item):
        pos = self.find(item)
        if type(pos) == int:
            #deleted
            self.__delitem__(pos)
        else:
            return pos



    def __str__(self): #for print magic method
        #[1,2,3]
        result = ''
        for i in range(self.n):
            result = result + str(self.A[i]) + ',' 

        return '[' + result[:-1] + ']'
           
```
