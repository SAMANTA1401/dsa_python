```python
class Node:
    def __init__(self, value):
        self.value = value
        self.next = None
```

```python
class Queue:
    def __init__(self):
        self.front = None
        self.near = None
    

    def is_empty(self):
        return self.front is None
    
    def size(self):
        temp = self.front
        count = 0
        while temp != None:
            temp = temp.next
            count += 1
        return count

    def enqueue(self, value):
        new_node = Node(value)
        if self.front == None:
            self.front = new_node
            self.rear = self.front
        else:
            self.rear.next = new_node
            self.rear = new_node
        print(self.rear.value)

    def dequeue(self):
        if self.front == None:
            return "Empty"
        else:
            temp = self.front
            self.front = self.front.next
        return temp.value
    
    def front_item(self):
        if self.front == None:
            return "Empty"
        else:
            return self.front.value
        
    def rear_item(self):
        if self.front == None:
            return "Empty"
        else:
            return self.rear.value
    

    def traverse(self):
        temp = self.front
        while temp != None:
            print(temp.value)
            temp = temp.next
            
```
